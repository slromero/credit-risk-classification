```python
# Import the modules
import numpy as np
import pandas as pd
import seaborn as sns
from pathlib import Path

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample

# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
lending_data_df = pd.read_csv("Resources/lending_data.csv", low_memory=False)

# Review the DataFrame
lending_data_df.head()
```

```python
lending_data_df['loan_status'].value_counts()
```

The value counts indicate that we have significantly fewer values for high-risk loans than we do for healthy loans.

### Fixing Class Imbalance with Undersampling

```python
# Split the dataframe into the majority and minority classes
minority_class = lending_data_df['loan_status'].value_counts().idxmin()
majority_df = lending_data_df[lending_data_df['loan_status'] != minority_class]
minority_df = lending_data_df[lending_data_df['loan_status'] == minority_class]

# Undersample the majority class to match the number of samples in the minority class
undersampled_majority = resample(majority_df, replace=False, n_samples=len(minority_df), random_state=42)

# Combine the undersampled majority class with the minority class and shuffle the rows
balanced_df = pd.concat([undersampled_majority, minority_df]).sample(frac=1, random_state=42)
balanced_df['loan_status'].value_counts()
```

### Checking for Highly Interdependent Features

```python
numeric_cols = balanced_df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr = numeric_cols.corr()

# Generate a heatmap
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True);
```

### Step 2: Create the labels set (`y`)  from the loan_status column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features
# Separate the y variable, the labels
y = balanced_df['loan_status'].ravel()

# Separate the X variable, the features
X = balanced_df.drop(columns='loan_status')
```

```python
# Review the y variable Series
y[:5]
```

```python
# Review the X variable DataFrame
X.shape
```

### Step 3: Split the data into training and testing datasets by using `train_test_split`.

```python
# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Instantiate a StandardScaler instance
scaler = StandardScaler()

# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
X_test_scaled
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(solver='lbfgs', random_state=1)

# Fit the model using training data
classifier.fit(X_train_scaled, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python
# Make a prediction using the training data
predictions_TRAIN = classifier.predict(X_train_scaled)
pd.DataFrame({"Prediction TRAIN": predictions_TRAIN, "Actual": y_train})
```

```python
# Make a prediction using the testing data
predictions = classifier.predict(X_test_scaled)
pd.DataFrame({"Prediction TEST": predictions, "Actual": y_test})
```

### Step 3: Evaluate the models performance by doing the following:

* Generate a confusion matrix.

* Print the classification report.

```python
# Generate a confusion matrix for the model TRAINING
confusion_matrix(y_train,predictions_TRAIN)
```

```python
# Print the classification report for the model
target_names = ["Healthy Loan TRAIN","High-Risk Loan TRAIN"]
print(classification_report(y_train, predictions_TRAIN, target_names = target_names))
```

```python
# Generate a confusion matrix for the model TEST
confusion_matrix(y_test,predictions)
```

```python
# Print the classification report for the model
target_names = ["Healthy Loan TEST","High-Risk Loan TEST"]
print(classification_report(y_test, predictions, target_names = target_names))
```

## Instantiate an K Nearest Neighbor Classifier instance.

```python
# Instantiate the KNeighborsClassifier model with n_neighbors = 3 
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
```

```python
# Create predictions using the testing data
y_pred_knn = knn.predict(X_test_scaled)
```

```python
# Print the classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred_knn))
```
